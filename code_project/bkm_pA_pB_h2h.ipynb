```python
# Création des 2 colonnes "player_A" / "player_B" et de la colonnes "target" (0 pour "A gagne", 1 pour "B gagne")

import pandas as pd
import numpy as np

df = pd.read_csv("csv_files/atp_data_4.csv", sep = ";")
df = df.drop(['Unnamed: 0'], axis=1)

# on fige la "seed" de random numpy pour reproduire le même aléatoire à chaque regénération de df
np.random.seed(33)

# on créer une colonne avec des 1 et des 2 au hasard
df["target"] = np.random.choice([0, 1], df.shape[0])

# on crée les colonnes joueur A et joueur B de manière 100% aléatoire
df["player_A"] = np.where(df['target'] == 1, df["Winner"], df["Loser"])
df["player_B"] = np.where(df['player_A'] == df["Winner"], df["Loser"], df["Winner"])
    
df
```

```python
################################################################
# variables head-to-head (h2h)
################################################################
```

```python
# Sélection des colonnes nécessaires pour la création des variables de type "head-2-head"
dfh = df[["id_week","Year_week","Tournament","Surface","Comment","player_A","player_B","Winner"]]
```

```python
# Création d'un identifiant unique pour chaque head-2-head (variable "id_h2h") avec tri par ordre alphabétique de type :
# "Federer R. - Nadal R." qui ne change jamais pour un head-2-head donnée, quels que soient les joueurs A et B de 
# la ligne observée
id_h2h_list = []
for i in range (0,58664):
    df1 = dfh.iloc[i:(i+1),:]
    player_A = df1.iloc[0]["player_A"]
    player_B = df1.iloc[0]["player_B"]
    list_a_and_b = [player_A, player_B]
    list_a_and_b.sort()
    id_h2h = str(list_a_and_b[0]) + " - " + str(list_a_and_b[1])
    id_h2h_list.append(id_h2h)
    
dfh["id_h2h"] = id_h2h_list   
```

```python
# Créations des variables "h2h_player_1" et "h2h_player_1" issues de "id_h2h"
dfh["h2h_player_1"] = dfh["id_h2h"].apply(lambda x: str(x).split(" -")[0])
dfh["h2h_player_2"] = dfh["id_h2h"].apply(lambda x: str(x).split(" -")[1]) 
```

```python
# Création des variables "h2h_player_1_win" et "h2h_player_2_win" qui valent 1 ou 0 selon qui remporte le match.
# Remarque : on ne prend pas en compte les victoires par forfait ("Comment" = "Walkover").
dfh["h2h_player_1_win"] = np.where(dfh["Comment"] == "Walkover", 0,np.where(dfh["h2h_player_1"] == dfh["Winner"], 1, 0))
dfh["h2h_player_2_win"] = np.where(dfh["Comment"] == "Walkover", 0,np.where(dfh["h2h_player_1_win"] == 1, 0, 1))
```

```python
# Création des colonnes h2h_player_1_score et h2h_player_2_score qui comptent le score de chaque joueur dans le head-to-head
h2h_score_player_1_list = []
h2h_score_player_2_list = []

for i in range (0,58664):
    score_1 = int((dfh.loc[((dfh["id_h2h"] == dfh["id_h2h"][i]))&(dfh["id_week"] < (dfh["id_week"][i])),\
                           ["h2h_player_1_win"]].sum()).iloc[0])
    score_2 = int((dfh.loc[((dfh["id_h2h"] == dfh["id_h2h"][i]))&(dfh["id_week"] < (dfh["id_week"][i])),\
                           ["h2h_player_2_win"]].sum()).iloc[0])
    h2h_score_player_1_list.append(score_1)
    h2h_score_player_2_list.append(score_2)

dfh["h2h_player_1_score"] = h2h_score_player_1_list
dfh["h2h_player_2_score"] = h2h_score_player_2_list
```

```python
# Création d'une variable h2h_match_number qui s'incrémente à chaque nouveau match dans le head-to-head (toutes surfaces
# confondues).
dfh["h2h_match_number"] = 1 + dfh["h2h_player_1_score"] + dfh["h2h_player_2_score"] 
```

```python
# Création des variables "h2h_player_A_score" et "h2h_player_B_score" issues de "h2h_player_1_score" et "h2h_player_2_score"
dfh["h2h_player_A_score"] = np.where(dfh["player_A"] == dfh["h2h_player_1"], dfh["h2h_player_1_score"], \
                                     dfh["h2h_player_2_score"])
dfh["h2h_player_B_score"] = np.where(dfh["player_B"] == dfh["h2h_player_1"], dfh["h2h_player_1_score"], \
                                     dfh["h2h_player_2_score"])
```

```python
# Création des colonnes h2h_score_last_5_player_1 et h2h_score_last_5_player_2 qui comptent le score de chaque joueur 
# dans le head-to-head
h2h_player_1_score_last_5_list = []
h2h_player_2_score_last_5_list = []

for i in range (0,58664):    
    score_1 = int((dfh.loc[((dfh["id_h2h"] == dfh["id_h2h"][i]))&(dfh["id_week"] < (dfh["id_week"][i]))\
                           &(dfh["h2h_match_number"] < (dfh["h2h_match_number"][i]))\
                           &(dfh["h2h_match_number"] >= ((dfh["h2h_match_number"][i]-5))),["h2h_player_1_win"]].sum()).iloc[0])
    score_2 = int((dfh.loc[((dfh["id_h2h"] == dfh["id_h2h"][i]))&(dfh["id_week"] < (dfh["id_week"][i]))\
                           &(dfh["h2h_match_number"] < (dfh["h2h_match_number"][i]))\
                           &(dfh["h2h_match_number"] >= ((dfh["h2h_match_number"][i]-5))),["h2h_player_2_win"]].sum()).iloc[0])
    h2h_player_1_score_last_5_list.append(score_1)
    h2h_player_2_score_last_5_list.append(score_2)

dfh["h2h_player_1_score_last_5"] = h2h_player_1_score_last_5_list
dfh["h2h_player_2_score_last_5"] = h2h_player_2_score_last_5_list
```

```python
# Création des variables "h2h_player_A_score_last_5" et "h2h_player_B_score_last_5" issues de "h2h_player_1_score_last_5" et 
# "h2h_player_2_score_last_5"
dfh["h2h_player_A_score_last_5"] = np.where(dfh["player_A"] == dfh["h2h_player_1"], dfh["h2h_player_1_score_last_5"], dfh["h2h_player_2_score_last_5"])
dfh["h2h_player_B_score_last_5"] = np.where(dfh["player_B"] == dfh["h2h_player_1"], dfh["h2h_player_1_score_last_5"], dfh["h2h_player_2_score_last_5"])
```

```python
# Création des colonnes h2h_player_1_score_surface et h2h_player_2_score_surface qui compte le score de chaque joueur 
# dans le head-to-head avec prise en compte de la surface
h2h_score_player_1_surface_list = []
h2h_score_player_2_surface_list = []

for i in range (0,58664):
    score_1 = int((dfh.loc[((dfh["id_h2h"] == dfh["id_h2h"][i]))&(dfh["id_week"] < (dfh["id_week"][i]))
                           &(dfh["Surface"] == dfh["Surface"][i]),["h2h_player_1_win"]].sum()).iloc[0])    
    score_2 = int((dfh.loc[((dfh["id_h2h"] == dfh["id_h2h"][i]))&(dfh["id_week"] < (dfh["id_week"][i]))
                           &(dfh["Surface"] == dfh["Surface"][i]),["h2h_player_2_win"]].sum()).iloc[0])                           

    h2h_score_player_1_surface_list.append(score_1)
    h2h_score_player_2_surface_list.append(score_2)

dfh["h2h_player_1_score_surface"] = h2h_score_player_1_surface_list
dfh["h2h_player_2_score_surface"] = h2h_score_player_2_surface_list
```

```python
# Création des variables "h2h_player_A_score_surface" et "h2h_player_B_score_surface" issues de 
# "h2h_player_1_score_surface" et "h2h_player_2_score_surface" 
dfh["h2h_player_A_score_surface"] = np.where(dfh["player_A"] == dfh["h2h_player_1"], dfh["h2h_player_1_score_surface"], \
                                     dfh["h2h_player_2_score_surface"])
dfh["h2h_player_B_score_surface"] = np.where(dfh["player_B"] == dfh["h2h_player_1"], dfh["h2h_player_1_score_surface"], \
                                     dfh["h2h_player_2_score_surface"])
```

```python
# Création d'une variable h2h_match_number_surface qui s'incrémente à chaque nouveau match dans le head-to-head sur
# la surface concernée par le match de la ligne
dfh["h2h_match_number_surface"] = 1 + dfh["h2h_player_1_score_surface"] + dfh["h2h_player_2_score_surface"] 
```

```python
# Création des colonnes h2h_player_1_score_last_5_surface et h2h_player_2_score_last_5_surface qui comptent le score 
# de chaque joueur dans le head-to-head avec prise en compte de la surface
h2h_player_1_score_last_5_surface_list = []
h2h_player_2_score_last_5_surface_list = []

for i in range (0,58664):    
    score_1 = int((dfh.loc[((dfh["id_h2h"] == dfh["id_h2h"][i]))&(dfh["id_week"] < (dfh["id_week"][i]))\
                           &(dfh["h2h_match_number_surface"] < (dfh["h2h_match_number_surface"][i]))\
                           &(dfh["h2h_match_number_surface"] >= ((dfh["h2h_match_number_surface"][i]-5)))\
                           &(dfh["Surface"] == dfh["Surface"][i]),["h2h_player_1_win"]].sum()).iloc[0])
    score_2 = int((dfh.loc[((dfh["id_h2h"] == dfh["id_h2h"][i]))&(dfh["id_week"] < (dfh["id_week"][i]))\
                           &(dfh["h2h_match_number_surface"] < (dfh["h2h_match_number_surface"][i]))\
                           &(dfh["h2h_match_number_surface"] >= ((dfh["h2h_match_number_surface"][i]-5)))\
                           &(dfh["Surface"] == dfh["Surface"][i]),["h2h_player_2_win"]].sum()).iloc[0])

    h2h_player_1_score_last_5_surface_list.append(score_1)
    h2h_player_2_score_last_5_surface_list.append(score_2)

dfh["h2h_player_1_score_last_5_surface"] = h2h_player_1_score_last_5_surface_list
dfh["h2h_player_2_score_last_5_surface"] = h2h_player_2_score_last_5_surface_list
```

```python
# Création des variables "h2h_player_A_score_last_5" et "h2h_player_B_score_last_5" issues de "h2h_player_1_score_last_5" et 
# "h2h_player_2_score_last_5"
dfh["h2h_player_A_score_last_5_surface"] = np.where(dfh["player_A"] == dfh["h2h_player_1"], \
                                                    dfh["h2h_player_1_score_last_5_surface"], dfh["h2h_player_2_score_last_5_surface"])
dfh["h2h_player_B_score_last_5_surface"] = np.where(dfh["player_B"] == dfh["h2h_player_1"], \
                                                    dfh["h2h_player_1_score_last_5_surface"], dfh["h2h_player_2_score_last_5_surface"])
```

```python
# on ajoute à df les 8 nouvelles variables :
# "h2h_player_A_score" et "h2h_player_B_score"
# "h2h_player_A_score_last_5" et "h2h_player_B_score_last_5"
# "h2h_player_A_score_surface" et "h2h_player_B_score_surface"
# "h2h_player_A_score_last_5_surface", "h2h_player_B_score_last_5_surface"
df["h2h_player_A_score"] = dfh["h2h_player_A_score"]
df["h2h_player_B_score"] = dfh["h2h_player_B_score"]
df["h2h_player_A_score_last_5"] = dfh["h2h_player_A_score_last_5"]
df["h2h_player_B_score_last_5"] = dfh["h2h_player_B_score_last_5"]
df["h2h_player_A_score_surface"] = dfh["h2h_player_A_score_surface"]
df["h2h_player_B_score_surface"] = dfh["h2h_player_B_score_surface"]
df["h2h_player_A_score_last_5_surface"] = dfh["h2h_player_A_score_last_5_surface"]
df["h2h_player_B_score_last_5_surface"] = dfh["h2h_player_B_score_last_5_surface"]
```
