<center><h1>  Peut-on faire confiance au bookmaker? </h1></center>

<br/>

En analysant la colonne **B365W** dans le dataset initial, on trouve que le bookmaker a prédit:
>
> + correctement le joueur gagnant (22306 fois). 
>
> + incorrectement le joueur gagnant (9212 fois). 
>
> + aucune décision au niveau du gagnant/perdant (673 fois).

supposons qu'un parrainage sera effectuer sur les matchs prévus gagnant par le bookmaker "**en éliminant les cas lorsqu'on a les mêmes cotes pour les deux joueurs**", et en mettant 1 euros sur chaque match, donc:

<p style="text-align: center;"> <ins> Résultat = somme des cotes gagnants - (nbr des cotes gagnants + nbr de cotes perdues) </ins>

<br/>

le résultat est: <span style="color: blue"> **-1717.73 euros** </span> sur 31518 matchs

> Nous sommes donc face au plusieurs possibilités :
> + Le Bookmaker a fait une erreur en prédisant le gagnant.
> + Le bookmaker a prédit le gagnant correctement, mais avec une marge bénéficiaire qui est en sa faveur.
> + Le bookmaker a prédit le gagnant correctement, mais il a remplacé le cote du gagnant par le cote du perdant pour tromper le parieur
    
Dans la suite nous allons créer un algorithme qui prédit le joueur gagnant, puis nous allons comparer le résultat de l'algorithme avec le résultat du bookmaker **B365**

<center><h1> Prédire le gagnant d'un match </h1></center>

<br/>

Dans cette partie on va utiliser les variables du dataset initial pour prédire le gagnant d'un match, pour cela on va choisir seulement les variables qui sont disponibles lors de la prediction.

> + Les variables utilisées dans l'algorithme sont: **Tournament, Series, Court, Surface, Winner, Loser, WRank, LRank, elo_winner, elo_loser, proba_elo**

#### Les étapes de l'algorithme

<br/>

> la méthode de la prédiction du joueur gagnant passe par des plusieurs étapes:
>
> + **étape1**: `On suppose que le joueur de la ligne target a joué n match au paravant donc le nouveau dataset contient n lignes. On lance l'agorithme sur j lignes "ensemble d'entrainement formant 80% du dataset trouvé" et on fait une prédiction sur la ligne numéros j+1 en sauvgardant le résultat trouvé dans une liste qui s'appelle prédiction.`
>
>
> + **etape2**: `je colle la ligne numéros j+1 avec ces informations  au dataset trouvé en haut et je relance de nouveau l'algorithme sur les 80% des observations pour faire une prediction sur la ligne j+2  et je sauvgarde le resultat trouvé dans la liste prédiction.`
>
>
> + **etape3**: `On repète les étapes sur les lignes restantes du dataset trouvé pour prédire à la fin le status du joueur sur la ligne target "s'il va gagner ou perdre le match"`
>
>
> + **dernière étape**: `On compare les résultats dans la liste de prédiction avec les résutats réels pour trouver un seuil de fidélité de la prediction au niveau de la ligne 'target'. par exemple: si l'algorithme a bien predit le joueur gagnant au niveau de la ligne 'target' <<c.à.d il a prédit que le joueur winner du dataset a gagné le match>>, danc ce cas on prendrons en considération le seuil de fidelité, s'il était supérieur à 70% par exemple cela signifie qu'il y a une grande confiance dans cette prévision.`


#### Exemple
> On lance l'algorithme sur la ligne d'indice 32400 en utilisant le classifeur KNN, l'algorithme a bien prédit le gagnant de ce match qui est **Haas T.** avec un seuil de fidélité de 81%
