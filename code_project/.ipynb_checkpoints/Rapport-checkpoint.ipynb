<center><h1> Prdire le gagnant d'un match </h1></center>

<br/>

Dans cette partie on va utiliser les variables du dataset initial pour prdire le gagnant d'un match, pour cela on va choisir seulement les variables qui sont disponibles lors de la prediction.

> + Les variables utilises dans l'algorithme sont: **Tournament, Series, Court, Surface, Winner, Loser, WRank, LRank, elo_winner, elo_loser, proba_elo**

#### Les tapes de l'algorithme

<br/>

> la mthode de la prdiction du joueur gagnant passe par des plusieurs tapes:
>
> + **tape1**: `On suppose que le joueur de la ligne target a jou n match au paravant donc le nouveau dataset contient n lignes. On lance l'agorithme sur j lignes "ensemble d'entrainement formant 80% du dataset trouv" et on fait une prdiction sur la ligne numros j+1 en sauvgardant le rsultat trouv dans une liste qui s'appelle prdiction.`
>
>
> + **etape2**: `je colle la ligne numros j+1 avec ces informations  au dataset trouv en haut et je relance de nouveau l'algorithme sur les 80% des observations pour faire une prediction sur la ligne j+2  et je sauvgarde le resultat trouv dans la liste prdiction.`
>
>
> + **etape3**: `On repte les tapes sur les lignes restantes du dataset trouv pour prdire  la fin le status du joueur sur la ligne target "s'il va gagner ou perdre le match"`
>
>
> + **dernire tape**: `On compare les rsultats dans la liste de prdiction avec les rsutats rels pour trouver un seuil de fidlit de la prediction au niveau de la ligne 'target'. par exemple: si l'algorithme a bien predit le joueur gagnant au niveau de la ligne 'target' <<c..d il a prdit que le joueur winner du dataset a gagn le match>>, danc ce cas on prendrons en considration le seuil de fidelit, s'il tait suprieur  70% par exemple cela signifie qu'il y a une grande confiance dans cette prvision.`


#### Exemple
> On lance l'algorithme sur la ligne d'indice 32400 en utilisant le classifeur KNN, l'algorithme a bien prdit le gagnant de ce match qui est **Haas T.** avec un seuil de fidlit de 81%
