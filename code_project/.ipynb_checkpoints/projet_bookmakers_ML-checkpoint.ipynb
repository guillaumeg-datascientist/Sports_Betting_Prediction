```python
df = pd.read_csv('atp_data_only_var.csv', sep = ";")
df = df.drop('Index', axis = 1)
```

```python
data = df.drop('target', axis = 1)
target = df['target']
```

```python
#X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state = 42)

X_train =  data.loc[:41772,:]
y_train =  target.loc[:41772]
X_test = data.loc[41772:,:]
y_test = target.loc[41772:]
```

```python
rf = ensemble.RandomForestClassifier()

grid_rf = model_selection.GridSearchCV(estimator=rf, param_grid={'max_features': ["sqrt", "log2", None ], 'min_samples_split': range(2, 40, 2)})
grid_rf.fit(X_train, y_train)

print('Meilleurs paramètres :', grid_rf.best_params_)

print("Taux de matchs bien prédits :", grid_rf.score(X_test,y_test))

y_pred = grid_rf.predict(X_test)
pd.crosstab(y_test, y_pred, rownames = ['Classe réelle'], colnames = ['Classe prédite par rf'])
```
